## ðŸŒ  Example 1: Planet Orbit

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
a = 1.0   # semi-major axis (AU)
theta = np.linspace(0, 2*np.pi, 500)  # angles

# Orbit coordinates (circle)
x = a * np.cos(theta)
y = a * np.sin(theta)

# Plot
plt.figure(figsize=(6,6))
plt.plot(x, y, 'b')             # orbit path
plt.plot(0, 0, 'yo', markersize=12, label="Star")  # star at center
plt.plot(a, 0, 'ro', label="Planet")               # planet at one position
plt.xlabel("X (AU)")
plt.ylabel("Y (AU)")
plt.title("Simple Planet Orbit")
plt.legend()
plt.axis("equal")
plt.show()
```

## ðŸŒ  Example 2: Exoplanet Transit Light Curve

```python
import numpy as np
import matplotlib.pyplot as plt

# Time (days)
time = np.linspace(0, 10, 500)

# Star brightness (normalized to 1)
flux = np.ones_like(time)

# Add a dip between day 4 and 6 (planet transit)
flux[(time > 4) & (time < 6)] -= 0.02  # 2% dip

# Plot
plt.plot(time, flux, 'k-')
plt.xlabel("Time (days)")
plt.ylabel("Normalized Brightness")
plt.title("Simulated Exoplanet Transit Light Curve")
plt.ylim(0.95, 1.01)
plt.show()
```

## ðŸŒ  Example 3: Hubbleâ€™s Law

```python
import numpy as np
import matplotlib.pyplot as plt

# Example distances (Mpc)
distance = np.linspace(0, 1000, 50)

# Hubble constant (km/s/Mpc)
H0 = 70  

# Velocity (km/s)
velocity = H0 * distance

# Plot
plt.scatter(distance, velocity, c='blue', label="Galaxies")
plt.plot(distance, velocity, 'r--', label=f"v = H0 * d (H0={H0})")
plt.xlabel("Distance (Mpc)")
plt.ylabel("Velocity (km/s)")
plt.title("Hubble's Law: Expansion of the Universe")
plt.legend()
plt.show()
```
